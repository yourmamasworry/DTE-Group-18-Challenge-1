```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
```

```python
## importing the kenyan population dataset

data_kenya= pd.read_csv( 'Kenyan population data.csv')
```

###### <span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#0e92ea"> According to unicef, 59% of Kenyans have access to safe drinking water. According to the world bank report dated December 6, 2018, Kenya has access rate of 75% to electricity.

 #### <span style="color:red;">The following section of EDA will use the percentage figures above and known population figures from KNBS to compute what proportion of Kenyans in either rural or urban areas have access to water and electricity and is therefore heavily subjective.</span>
 #### It is also assumed that the rates remain constant over time.


```python
print(data_kenya)
```

```python
print(data_kenya.columns)
```

```python
# fitting the percentages in variables.
water_access= 0.25
electricity_access= .75
data_kenya['Rural access to water'] = data_kenya['Rural population'] * water_access
data_kenya['Urban access to water'] = data_kenya['Urban population'] * water_access
data_kenya['Rural access to electricity'] = data_kenya['Rural population'] * electricity_access
data_kenya['Urban access to electricity'] = data_kenya['Urban population'] * electricity_access


# Print the updated DataFrame
pd.options.display.float_format = '{:.2f}'.format

# Print the updated DataFrame
print(data_kenya)
```

#### bar plots to display access to both water and electricity

```python
# Extract relevant data for the plot
years = data_kenya['Year']
rural_water = data_kenya['Rural access to water']
urban_water = data_kenya['Urban access to water']
rural_electricity = data_kenya['Rural access to electricity']
urban_electricity = data_kenya['Urban access to electricity']


# Set the width of the bars
bar_width = 0.35
indices = np.arange(len(years))

# Create the clustered bar chart
plt.figure(figsize=(12, 6))
plt.bar(indices - bar_width, rural_water, bar_width, label='Rural Water')
plt.bar(indices - bar_width, rural_electricity, bar_width, label='Rural Electricity', alpha=0.5, align='edge')
plt.bar(indices + bar_width, urban_water, bar_width, label='Urban Water')
plt.bar(indices + bar_width, urban_electricity, bar_width, label='Urban Electricity', alpha=0.5, align='edge')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Access Count')
plt.title('Access to Water and Electricity (Rural vs. Urban) Over Time')
plt.xticks(indices, years, rotation=45)

# Add a legend
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
